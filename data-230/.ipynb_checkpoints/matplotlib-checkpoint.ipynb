```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10)

y1 = x
y2 = x+1
y3 = x+2
```

```python
plt.plot(x, y1, label='x', color="blue", linestyle="solid")
plt.plot(x, y2, label="x+1", color="orange", linestyle="dashed")
plt.plot(x, y3, label="x+3", color="green", linestyle="dotted")

plt.xlabel("X")
plt.ylabel("Value")

plt.title('Linear Function')


plt.grid(True)

```
