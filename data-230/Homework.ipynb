```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(-2, 2, 100)
x1 = np.linspace(0.01, 2, 100)

y1 = x * 2
y2 = x ** 2
y3 = np.log(x1)


plt.plot(x, y1, label='y1', color="blue", linestyle="solid")
plt.plot(x1, y2, label='y2', color="red", linestyle="solid")
plt.plot(x1, y3, label='y3', color="green", linestyle="solid")

plt.title("Assignment 0")
plt.xlabel("X")
plt.ylabel("Y")

plt.grid(True)


```
