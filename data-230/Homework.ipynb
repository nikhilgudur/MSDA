```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(-2, 2, 100)
x1 = np.linspace(0.01, 2, 100)

# 2x
y1 = x * 2

# x^2 
y2 = x ** 2

# log(x)
y3 = np.log(x1)


plt.plot(x, y1, label='x*2', color="blue", linestyle="solid")

plt.plot(x, y2, label='x^2', color="red", linestyle="dashed")

plt.plot(x1, y3, label='log(x)', color="green", linestyle="solid")

plt.title("Assignment 1")
plt.xlabel("X")
plt.ylabel("Y")

plt.grid(True)

plt.show()


```
