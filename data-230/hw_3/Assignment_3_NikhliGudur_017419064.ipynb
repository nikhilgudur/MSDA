```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# Question 1

diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='cut', y='price', data=diamonds)
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Price by Cut')
plt.show()
```

#### Observations from the diamonds dataset by the cut of the diamonds

* Median prices of diamonds with Premium cut are higher than the other cuts. Followed by Fair and Good.
* Premium cut diamonds have the highest price, followed by Very Good and Good cuts.
* Q3 is higher for Premium cut diamonds, followed by Very Good and Fair cuts.
* There are many outliers in the price of diamonds with Ideal cut.

```python
sns.boxplot(y='carat', x='color', data=diamonds)
plt.xlabel('Color')
plt.ylabel('Carat')
plt.title('Carat by Color')
plt.show()
```

#### Observations for diamonds for carat and their color:

* Diamonds with color J have the highest median price, followed by I and H.
* Diamonds with color D have the highest Q3, followed by E and D.
* Diamonds with color J have the highest Q1, followed by I and H.
* J color diamond has the highest number of caret, ~5

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Question 2

sns.violinplot(x='class', y='fare', data=titanic)
plt.xlabel('Class')
plt.ylabel('Fare')
plt.title('Fare by Class')
plt.show()
```

#### Observations of Titanic dataset with fair by class:

* First class passengers paid the highest fares on average. The median fare for first class passengers was about 72.5, while the median fare for second class passengers was about 13 and the median fare for third class passengers was about 7.75.
* There was a wider range of fair prices paid by first class passengers than by second or third class passengers. This is indicated by the wider bodies of the violins at the higher price points.
* There were a few outliers in each class who paid very high fares. For example, the highest fare paid by a first class passenger was more than 500, the highest fare paid by a second class passenger was around 26, and the highest fare paid by a third class passenger was around 43.5.


```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Question 3

sm.qqplot(tips['total_bill'], line="s")
plt.title('Total Bill')
plt.show()
```

#### Observations of Total bills from the Tips dataset 
The total bill amounts do not seem to follow a normal distribution. The slope of the line is very steep, which suggests that the total bill amounts are not normally distributed. Additionally, the points do not fall along a straight line, but rather curve upwards, which is another indication of non-normality.
