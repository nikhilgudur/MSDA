```python
import numpy as np
import numpy.linalg as nla
from skimage import io
from scipy.linalg import svd as SVD
import matplotlib.pyplot as plt
```

Find eigenvalues and Eigenvector of the matrix A = [1 4][-4 -7]

```python
M = np.array([[1, 4], [-4, -7]])
evalue, evector = nla.eig(M)

print("Eigen Value:\n", evalue)
print("Eigen Vector:\n", evector)
```
