```python
# Step 1
# Importing Moonlanding.csv into a dataframe using pandas
import pandas as pd


df = pd.read_csv('Moonlanding.csv')
```

```python
# Step 2
# To show how much data we have in the dataframe
print(len(df))
```

```python
# Step 3

# Head of the dataframe
print(df.head())

# Data from a particular row
print(df.loc[10])

# Data from multiple rows
print(df.loc[[99, 100, 101]])

# Shape of the dataframe
print(df.shape)

print(df.columns)
```

```python
# contingency table
pd.crosstab(index=df["Operator"], columns=df["Outcome"], margins=False)
```

```python
# Proportions of Mission Outcome by Mission Type
stats = pd.crosstab(df["Outcome"], df["Mission Type"])

proportions = stats/len(df)
print(proportions)
```
