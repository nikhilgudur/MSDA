```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
```

```python
# Question 1
# 1.A
a0 = 2
a1 = 3
a2 = 2

N  = 1000

x, y  = np.random.rand(N), np.random.rand(N)
noise = 0.01 * np.random.randn(N)

z = a0 +  a1*x + a2*y + noise

plt.scatter(x, y, z)
plt.show()
```

```python
# Question 1
# 1.B

X = np.column_stack((np.ones(N), x, y))
a = la.solve(X.T @ X, X.T @ z)

a0 = a[0]
a1 = a[1]
a2 = a[2]

print("a0 = ", a0, "a1 = ", a1, "a2 = ", a2)
```

```python
# Question 2

v1 = np.array([1, -2, 3])
v2 = np.array([-2, 4, -6])
v3 = np.array([3, -6, 9])

A = np.column_stack((v1, v2, v3))

rank = np.linalg.matrix_rank(A)

if rank == 3:
  print("The vectors are linearly independent")
else:
  print("The vectors are linearly dependent")

```

```python
# Question 3
# 3.A.1

v1 = np.array([1, 2, 3])
v2 = np.array([2, 3, 4])

norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)

print("Normal form of", v1,"\n", norm_v1)

print("Normal form of", v2,"\n", norm_v2)

```

```python
# 3.A.2

dot_product = np.dot(v1, v2)
print("Dot product of", v1, "and", v2, "\n", dot_product)
```

```python
# 3.B.1

matrix = np.array([[2, 1, 3], [0, 1, 4], [1, 2, 5]])

print("Shape of the Matrix\n", matrix.shape, "\n")


# 3.B.2

# transpose = matrix.T
transpose = np.transpose(matrix)

print("Matrix\n", matrix, "\n")
print("Transpose of the Matrix\n", transpose, "\n")
print("Inverse of the Matrix\n", la.inv(matrix), "\n")
print("Inverse of the Transpose of the Matrix\n", la.inv(matrix).T, "\n")

matrix_rank = np.linalg.matrix_rank(matrix)

print("Rank of the Matrix\n", matrix_rank)
```
