```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./retail_sales_dataset.csv')
```

```python
# Descriptive statistics of the dataset
df.describe()
```

```python
# Histogram of Total Amount

histogram = df["Total Amount"].hist(bins=10)
```

```python
# Descriptive Statistics of Male customers with product category of Clothing

male_customers = df[(df["Gender"] == "Male") & (df['Product Category'] == "Clothing")]
male_customers.describe()

```

```python
# Descriptive Statistics of Female customers with product category of Clothing

female_customers = df[(df["Gender"] == "Female") & (df["Product Category"] == "Clothing")]
female_customers.describe()

```

```python
plt.figure(figsize=(8,6))
plt.title("Distribution of Total Amount between males and females for clothing")
plt.grid(axis='y')
plt.xlabel("Clothing")
plt.ylabel("Amount")
data = [female_customers["Total Amount"][female_customers["Product Category"] == "Clothing"],
        male_customers["Total Amount"][male_customers["Product Category"] == "Clothing"]]

plt.boxplot(data, labels=['Female', 'Male'])
plt.show()
```
