```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./retail_sales_dataset.csv')
```

```python
df.describe()
```

```python
histo = df["Total Amount"].hist(bins=10)
```

```python
male_customers = df[(df["Gender"] == "Male")]


male_customers["Product Category"].hist(bins=10)
```

```python
female_customers = df[(df["Gender"] == "Female")]

female_customers['Product Category'].hist(bins=10)

```

```python
plt.figure(figsize=(8,6))
plt.title("Total Amount")
plt.grid(axis='y')
plt.xlabel("Clothing")
plt.ylabel("Amount")
data = [female_customers["Total Amount"][female_customers["Product Category"] == "Clothing"],
        male_customers["Total Amount"][male_customers["Product Category"] == "Clothing"]]

plt.boxplot(data, labels=['Female', 'Male'])
plt.show()
```
